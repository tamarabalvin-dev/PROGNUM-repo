```python
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 1.0243e+26, 8.6810e+25, 5.9736e+24,
          4.8685e+24, 6.4185e+23, 3.3022e+23, 7.349e+22, 1.25e22]
moonmass = masses[-2]
smallermass = []
for m in masses:
    if m <= moonmass:
        smallermass += [m]
print(f'All masses smaller or equal to the Moon: {smallermass}')

slicedlist = masses[6:]
print(f'The sliced list: {slicedlist}')

total = sum(slicedlist)
number = len(slicedlist)
averagemass = total/number
print(f'The average mass of the sliced list is {averagemass}')
```
