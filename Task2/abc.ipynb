```python
#Inputing variables
print(f'Variables:')
a = int(input('a = '))
b = int(input('b = '))
c = int(input('c = '))

#Computing discriminant
D = b**2 - 4*a*c

#Root determination
if D > 0:
    x1 = (-1*b + D**(1/2))/(2*a)
    x2 = (-1*b - D**(1/2))/(2*a)
    print(f'Roots to your quadratic equation: {x1} and {x2}')
elif D == 0:
    x = (-1*b)/(2*a)
    print(f'The root to your quadratic equation: {x}')
else:
    print(f'Your quadratic equation has no solutions.')

#For a = 1, b = 2, c = 1, solution is only -1.
```
