```python
import math

#Ask for user input

x = int(input('x = '))
y = int(input('y = '))

#Angle between x-axis and (xy) vector

angle = math.atan2(y,x)

#Evaluate quadrant

if 0 <= angle < math.pi/2:
    quad = 'first'
elif math.pi/2 <= angle < math.pi:
    quad = 'second'
elif -math.pi <= angle < -(math.pi/2):
    quad = 'third'
else:
    quad = 'fourth'
    
print('(x,y) belongs to the', quad, 'quadrant.')
```

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

#Conversion from list to set

set_night1 = set(night1)
set_night2 = set(night2)

#Finding intersection

intersection = list(set.intersection(set_night1, set_night2))
print(f'intersection of sets {intersection}')

#Using in operator

if 5 in intersection:
    print(f'5 is an element of the intersection list.')
else:
    print(f'5 is not an element of the intersection list.')
    
if 5 in night1:
    print(f'5 is a member of the set night1.')
else:
    print(f'5 is not a member of the set night1.')
```
